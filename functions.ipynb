**1. Simple function**

```python
def func():
  print("Hello, Welcome to python functions!")

func()
```

**2. Function with Parameters**

```python
def greet(name):
  print("Hello", name)

greet("Rohan")
```

**3. function with return value**

```python
def add(a, b):
  return a+b

print("Sum is",add(4,6))
```

**4. function with user input**

```python
def square():
  num = int(input("Enter a number: "))
  return(num*num)

print("Square is:",square())
```

**5. Default Arguments**

```python
def greet(name="Allen"):
  print("Hello", name)

greet()
greet("Bob")
```

**6. Keyword arguments**

```python
def user_info(name, age):
  print("Name: ", name)
  print("Age: ", age)

user_info(age=24, name='Chirag')
```

**7. Function retuning multiple values**

```python
def calculator(a, b):
  return a+b, a-b, a*b, a/b

sum_val,diff_val,pro_val,divide_val = calculator(2, 5)
print("Sum: ", sum_val)
print("Difference: ", diff_val)
print("Product: ", pro_val)
print("Division: ", divide_val)
```
