**1. Simple if statement**

```python
age = int(input("Enter your age: "))

if age >= 18:
    print("You are eligible to vote.")
```

**2. if-else Statement**

```python
num = int(input("Enter a number: "))
if(num % 2 == 0):
  print(num, "is even")
else:
  print(num, "is odd")
```

**3. if - elif - else Statement**

```python
marks = int(input("Enter your marks: "))

if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
elif marks >= 60:
    print("Grade C")
else:
    print("Fail")
```

**4. Nested if**

```python
age = int(input("Enter your age: "))
id = input("Do you have ID? (yes/no): ")

if age >= 18:
    if id == "yes":
        print("Entry allowed")
    else:
        print("ID required")
else:
    print("Underage")
```
