**1. Basic User Input**

```python
name = input("Enter your name: ")
sName = input("Enter your school name: ")
roll_no = input("Enter your Roll no.: ")

print("My name is", name, ", school name is", sName, ", and roll no. is", roll_no)
```

**2. Type Conversion**

```python
num1 = int(input("Enter a number: "))
num2 = int(input("Enter other number: "))
sum = num1+num2
print("Sum of given numbers is:",sum)
```

**3. Using F-strings**

```python
city = input("Enter Your City: ")
print(f"You live in {city}")
```

**4. Multiple inputs in one line**

```python
a, b, c = input("Enter the marks of all three subjects separated by space: ").split()

a = int(a)
b = int(b)
c = int(c)

percentage = (a+b+c)/3

print("Percentage: ",percentage)
```
