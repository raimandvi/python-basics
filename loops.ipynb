**1. For Loop**

```python
num = int(input("Enter a number: "))
for i in range(1, 11):

  print(num, "*", i,  "=",  num * i)
```

**2. While Loop**

```python
i = 1
while i <= 8:
    print(i)
    i += 1
```

**3. Loop with break**

```python
for i in range(1, 10):
    if i == 5:
        break
    print(i)
```

**4. Loop with continue**

```python
for i in range(1, 6):
    if i == 3:
        continue
    print(i)
```
